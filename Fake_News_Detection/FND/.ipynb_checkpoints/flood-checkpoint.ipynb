```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('FloodPrediction.csv')
```

```python
df.head(40)
```

```python
df.shape
```

```python
df.drop(['Station_Number', 'X_COR', 'Y_COR', 'LATITUDE', 'LONGITUDE', 'ALT', 'Period' ], axis='columns', inplace=True)
```

```python
df.head(5)
```

```python
df.describe()
```

```python
dict = {'Station_Names': 'Place',
        'Flood?': 'Flood'}

df.rename(columns=dict,
          inplace=True)
```

```python
df.head(5)
```

```python
df['Year']= df['Year'].astype('int')
df['Month']= df['Month'].astype('int')
df['Max_Temp']= df['Max_Temp'].astype('float')
df['Min_Temp']= df['Min_Temp'].astype('float')
df['Rainfall']= df['Rainfall'].astype('float')
df['Relative_Humidity']= df['Relative_Humidity'].astype('float')
df['Wind_Speed']= df['Wind_Speed'].astype('float')
df['Cloud_Coverage']= df['Cloud_Coverage'].astype('float')
```

```python
df.Flood.value_counts(normalize=True)
```

```python
df.isna().sum()
```

```python
df.fillna(0, inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```
